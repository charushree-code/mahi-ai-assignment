importing numpy,pandas,linearregression model


```
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

data is in form of y=2x1+3x2+4x3+5

```
data=pd.read_csv('data.csv')
```

```
data
```

```
x=data[['x1','x2','x3']].values
```

```
x
```

```
y=data[['y']].values
y
```

```
model=LinearRegression()
```

```
model.fit(x,y)
```

the intercept of the equation is 5

```
model.intercept_
```

the coefficients of x1=2,x2=3,x3=4

```
model.coef_
```

```
y_predict=model.predict(x)
```

```
y_predict
```

if x1=3,x2=2,x3=1 then y should be 21 according to the equation y=2x1+3x2+4x3+5

```
x_new=np.array([[3,  2,  1]])
```

```
y_new=model.predict(x_new)
```
